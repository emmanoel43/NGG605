```python
# Assuming 0.7% of the population has HIV
population = 1000
infection_rate= 0.007
hiv_positive = int(population * infection_rate)
hiv_negative = population - hiv_positive

# False positive rate (person without HIV tests positive)
false_positive_rate = 0.05
false_positives = int(hiv_negative * false_positive_rate)

# True positives (everyone with HIV tests positive)
true_positives = hiv_positive

# Total positive tests
total_positives = true_positives + false_positives

print("Exercise #2: Infection Rate 0.7%")
print("Number of people with HIV:", hiv_positive)
print("Number of people without HIV:", hiv_negative)
print("Number of false positives:", false_positives)
print("Number of true positives:", true_positives)
print("Total positive tests:", total_positives)

# Probability of true infection (Bayes' theorem)
true_positives = hiv_positive
false_positives = int(hiv_negative * false_positive_rate)
total_tests = true_positives + false_positives
probability_true_infection = true_positives / total_tests

print("Probability of true infection:", probability_true_infection)
print("*")

# Loop through infection rates from 0 to 1 in steps of 0.1
import numpy as np
for infection_rate in np.arange(0, 1.1, 0.1):
  hiv_positive = int(population * infection_rate)
  hiv_negative = population - hiv_positive

  # Calculate false positives and true positives
  false_positives = int(hiv_negative * false_positive_rate)
  true_positives = hiv_positive

  # Total positive tests
  total_positives = true_positives + false_positives

  # Probability of true infection (Bayes' theorem)
  probability_true_infection = true_positives / total_positives if total_positives > 0 else 0

  # Print results for the current infection rate
  print(f"Exercise #2: Infection Rate {infection_rate * 100:.1f}%")
  print("Number of people with HIV:", hiv_positive)
  print("Number of people without HIV:", hiv_negative)
  print("Number of false positives:", false_positives)
  print("Number of true positives:", true_positives)
  print("Total positive tests:", total_positives)
  print("Probability of true infection:", probability_true_infection)
  print("*")
```
